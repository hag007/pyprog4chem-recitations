#### Example #1: `if` vs `elif`

````python
# "Input" values 
rick_chips = 137
morty_chips = 731

if rick_chips > morty_chips:
    print("Rick won")
elif morty_chips > rick_chips:
    print("Morty won")
else:
    print("It's a tie!")
    
print("end")
````

````python
# "Input" values 
rick_chips = 137
morty_chips = 731

if rick_chips > morty_chips:
    print("Rick won")
if morty_chips > rick_chips:
    print("Morty won")
else:
    print("It's a tie!")
    
print("end")
````

#### Example #2: nested `if`

#### Write a function that returns whether a number is "positive even", "positive odd", "negative even", "negative odd" or "zero"

````python
# "Input" values 

def categorize_number(n):
    if n > 0:
        if n%2 == 0:
            return "positive even"
        else:
            return "positive odd"
    elif n < 0:
        if n%2 == 0:
            return "negative even"
        else:
            return "negative odd"
    else:
        return "zero"
````

````python
categorize_number(138)
categorize_number(-57)
categorize_number(0)
````

### Reminder: Exercise 1
- Return the number resulting from reversing the digit of a given <b>four-digit</b> number
- No `str` functions

Example: 1370 --> 731 (0731)

````python
num=1370

num1=(num%10)*1000 
num2=(num%100//10)*100 
num3=(num%1000//100)*10 
num4=((num)//1000)*1

print(num1,num2,num3,num4)
print(num1+num2+num3+num4)
````

#### Now with functions...

````python
def reverse_digits(num):
    num1=num%10*1000 
    num2=num//1000 
    num3=num%100//10*100 
    num4=num%1000//100*10 
    return num1+num2+num3+num4
````

````python
print(reverse_digits(1370))
print(reverse_digits(1375))
````

### Reminder: Exercise 2

#### Write a code that receives a string `s` with only lower letters, and prints a similar string, but with the middle letter in an upper case format.  

e.g.,  
ef**g**hi $\rightarrow$ ef<span style="color:red">**G**</span>hi   
efg**h**ij $\rightarrow$ efg<span style="color:red">**H**</span>ij  





#### Putting in all together in a function

````python
 def cap_in_middle(s):
    n = len(s)
    middle = n // 2
    middle_letter = s[middle]
    U_middle_letter = middle_letter.upper()
    return s[0:middle] + U_middle_letter + s[middle+1:]
    
````

````python
print(cap_in_middle('limerick'))
print(cap_in_middle('rickandmorty'))
````
